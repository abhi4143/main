# 2. Scarpping Tables from COVID19 Telangana State Wikipedia Page

```python
wiki_url = 'https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Telangana'


response = requests.get(wiki_url)
soup = BeautifulSoup(response.text,'html.parser')

table1 = soup.find('table',attrs={'class':'infobox'})
df = pd.read_html(str(table1))
```
